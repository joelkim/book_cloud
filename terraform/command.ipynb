# 테라폼 명령

## 테라폼 명령 실행 방법

- `terraform 커맨드 [인수 ...]` 형식으로 실행
- 주요 커맨드

  | 명령어 | 설명 |
  |-|---|
  | `terraform init`    | Terraform 구성 파일 디렉터리 초기화 (플러그인 설치, 모듈 설치, 백엔드 초기화) |
  | `terraform plan`    | 변경 사항 미리보기(실행 계획) 생성 |
  | `terraform apply`   | 변경 사항 적용 |
  | `terraform destroy` | 모든 리소스 삭제 (terraform apply -destroy의 별칭) |

## 테라폼 프로젝트 구축 순서

- 새로 만드는 테라폼 프로젝트는 다음과 같은 순서로 구축
  1. tf 파일 생성
  2. `terraform init` 명령으로 프로젝트 초기화
  3. `terraform plan` 명령으로 미리 보기 및 검토
  3. `terraform apply` 명령으로 클라우드에 적용


## 프로젝트 초기화

- 프로젝트용 tf 파일을 제작후 최초에 `terraform init` 명령
- 반드시 프로바이더 설정이 있어야 함
- 해당 프로바이더에 대한 바이너리를 `.terraform` 디렉토리 아래에 다운로드
- 프로젝트 디렉토리에 바이너리 검증용 해시코드가 저장된 `.terraform.lock.hcl` 파일 생성

```bash
$ tree .         
.
└── provider.tf

1 directory, 1 file

$ terraform init
Initializing the backend...
Initializing provider plugins...
- Finding hashicorp/aws versions matching "6.27.0"...
- Installing hashicorp/aws v6.27.0...
- Installed hashicorp/aws v6.27.0 (signed by HashiCorp)
Terraform has created a lock file .terraform.lock.hcl to record the provider
selections it made above. Include this file in your version control repository
so that Terraform can guarantee to make the same selections by default when
you run "terraform init" in the future.

Terraform has been successfully initialized!

You may now begin working with Terraform. Try running "terraform plan" to see
any changes that are required for your infrastructure. All Terraform commands
should now work.

If you ever set or change modules or backend configuration for Terraform,
rerun this command to reinitialize your working directory. If you forget, other
commands will detect it and remind you to do so if necessary.

$ tree -a
.
├── .terraform
│   └── providers
│       └── registry.terraform.io
│           └── hashicorp
│               └── aws
│                   └── 6.27.0
│                       └── darwin_amd64
│                           ├── LICENSE.txt
│                           └── terraform-provider-aws_v6.27.0_x5
├── .terraform.lock.hcl
└── provider.tf
```

## 미리보기

- `terraform plan` 명령으로 어떤 변경이 발생할지 미리 확인

```bash
$ terraform plan

No changes. Your infrastructure matches the configuration.

Terraform has compared your real infrastructure against your configuration and found no differences, so no changes are needed.
```
