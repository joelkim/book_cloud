# tf 파일

## tf 파일

- 테라폼의 설정 파일은 `.tf` 확장자 파일로 구성
- 하나의 디렉토리에 있는 모든 tf 파일은 하나로 합쳐서 인식
- 파일의 이름은 아무런 의미가 없음
- 선언형 언어를 사용하기 때문에 파일의 순서는 상관없음

## HCL 언어

- tf 파일은 HC(HashiCorp Configuration Language) 언어 사용
- 선언형 언어
- 블럭으로 구성

## 블럭

- 블럭: HCL 언어의 기본 요소
  - 블럭 타입
    - 블럭 타입을 나타내는 키워드 
  - 라벨(label)
    - 블럭의 주요 명세를 지정하는 인수
    - 블럭 타입에 따라 필요한 라벨의 개수가 달라짐 (0개 ~ 2개)
  - 인수(argument)
    - 해당 블럭의 속성
    - 인수 명세는 블럭과 블럭 라벨에 따라 달라짐
  - 하위 블럭
    - 블럭 내부의 블럭

```bash
블럭_타입 ["라벨1" "라벨2" ...] {
  인수 = 값
  하위_블럭 {
    속성 = 값
  }
}
```

## 블럭 타입

- 12가지 블럭 타입 제공
- 주요 블럭 타입

  | 블럭 타입 | 키워드 | 라벨 수 | 설명 |
  |-|-|-|-|
  | 테라폼 블럭 | terraform | 0 | 테라폼 프로젝트 정의 |
  | 프로바이더 블럭 | provider  | 1 (프로바이더 이름) | 프로바이더 |
  | 데이터소스 블럭 | data      | 2 (리소스 타입 + 리소스 이름) | 읽어올 수 있는 기존 리소스에 대한 정보 |
  | 리소스 블럭 | resource  | 2 (리소스 타입 + 리소스 이름) | 생성 또는 변경할 리소스에 대한 설정 |
  | 변수 블럭 | variable  | 1 | 변수 |


## terraform 블럭

- 프로젝트 자체에 대한 설정
- 대표적인 서브 블럭
  - required_version: 테라폼 자체의 버전
  - required_providers: 필요한 프로바이더 목록과 버전
  - backend: 상태를 저장할 장소 등에 대한 지정
- terraform 블럭과 서브 블럭은 생략 가능
  - 생략시 각각의 디폴트 설정에 따름

```bash
terraform { 
  required_providers { 
    aws = { 
    source = "hashicorp/aws"
    version = "~> 5.0"
  }
 }
}
```

## 프로바이더 블럭

- 클라우드 리소스를 제공하는 서비스 업체(AWS, GCP 등)에 대한 설정
- 프로바이더 이름을 지정하는 하나의 라벨 필요
- 프로바이더에 따라 브로바이더 블럭 자체의 필요 명세, 구현할 수 있는 리소스 블럭 등이 달라짐
- 세부 명세는 각 프로바이더 레지스트리 참조

## 테라폼 레지스트리

- 블럭의 세부 명세를 공개하는 웹 문서
- AWS 프로바이더 레지스트리: https://registry.terraform.io/providers/hashicorp/aws/latest/docs
- 주요 내용
  - 프로바이더 블럭의 세부 인수 
  - 사용할 수 있는 리소스 블럭과 데이터 소스블럭의 "리소스 타입" 목록
  - 각 "리소스 타입"의 인수 명세

## 데이터소스 블럭

- 현재 클라우드에 이미 만들어져 있는 리소스들에 대한 정보를 동적으로 읽어와 다른 블럭에서 사용할 수 있게 하는 블럭

   ```bash
   data 리소스_타입 리소스_이름 {
      인수1
      인수2
   }
   ```
- 다음 형식으로 다른 블럭에서 참조(사용) 가능

   ```bash
   data.리소스_타입.리소스_이름.인수
   ```
