# tf 파일

## tf 파일

- 테라폼은 다음 파일로 구성
  - tf 파일
- 하나의 디렉토리에 있는 모든 tf 파일은 하나로 합쳐서 인식
- 파일의 이름은 아무런 의미가 없음
- 선언형 언어를 사용하기 때문에 파일의 순서는 상관없음

## HCL 언어

- tf 파일은 HCL 언어 사용
- HashiCorp Configuration Language
- 선언형 언어
- 블럭으로 구성

## 블럭

- HCL 언어의 기본 요소
  - 블럭 타입
    - 블럭 타입을 나타내는 키워드 
  - 라벨
    - 블럭 타입에 따라 필요한 라벨의 개수가 달라짐 (0개 ~ 2개)
  - 속성
    - 해당 블럭의 속성
  - 하위 블럭
    - 블럭 내부의 블럭

```bash
블럭_타입 ["라벨1" "라벨2" ...] {
  속성 = 값
  하위_블럭 {
    속성 = 값
  }
}
```

## 블럭 타입

- 12가지 블럭 타입
- 그 중 많이 사용되는 블럭 타입

| 블럭 타입 | 라벨 수 | 설명 |
|-|-|-|
| terraform | 0 | 테라폼 프로젝트 정의 |
| provider  | 1 | 프로바이더 |
| resource  | 2 | (제어할) 리소스 |
| data      | 2 | (기존) 리소스 |
| variable  | 1 | 변수 |
