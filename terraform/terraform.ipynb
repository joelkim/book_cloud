# 테라폼이란

## 동작 원리

- 사용자는 테라폼에서 제공하는 언어로 클라우드를 구성하도록 프로젝트 제작
- 클라우드 서비스에서 제공하는 프로바이더 바이너리 파일이 테라폼 프로젝트 언어를 해석하여 자체적인 클라우드 API로 변경하여 클라우드 서비스에 적용

## 테라폼 설치

- 테라폼 명령 실행을 위한 바이너리 설치 필요
- `terraform` 또는 `terraform.exe` 형태의 단일 파일
- 커맨드라인 프로그램

## 테라폼 프로젝트 구성

- 사용자 구성 파일
  - tf 파일
- 자동 생성 파일
  - 초기 생성 
    - 프로바이더 파일
    - `.terraform.lock.hcl` 락 파일 
  - 적용시 생성
    - `terraform.tfstate` 상태 저장 파일
