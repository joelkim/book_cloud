# 가상 컴퓨터

## 인스턴스

### 인스턴스 가동

- EC2 > Instances > Launch an instance

- 가동시 설정 사항
  1. Name and Tags
  2. Application and OS Images (Amazon Machine Image)
  3. Instance type
  4. Key pair (login)
  5. Network settings
  6. Configure storage
  7. Advanced details

### 인스턴스 유형

- 명명규칙: 메인 카테고리 + 세대 + 옵션 카테고리
  - 예시: t4g = t(T 패밀리) + 4(4세대) + g(ARM64 기반)

- 메인 카테고리
  - T 패밀리: 저가형, 순간확장(burstable)
  - M 패밀리: 일반목적
  - C 패밀리: 고성능 CPU
  - R 패밀리: 대용량 메모리 
  - X 패밀리: 초대용량 메모리 
  - D 패밀리: 대용량 HDD 저장장치
  - I 패밀리: 대용량 SDD 저장장치
  - P, G, CG 패밀리: GPU 장착
  - F 패밀리: FPGA 장착

- 옵션 카테고리
  - g: ARM64 기반
  - d: 로컬 NVMe SSD
  - n: 네트워크/EBS 최적화

- vCPU 및 메모리 크기
  - nano
  - micro
  - small
  - large
  - x2large

## SSM(AWS Session Manager)

- ssh 등을 사용하여 외부에서 인스턴스로 직접 인바운드 접속을 하는 것이 아니라
- 인스턴스 내에 SSM 에이전트가 설치되어 SSM 서비스로 아웃바운드 접속하고
- 사용자는 SSM 서비스로 접속하여 SSM 서비스가 중간에서 릴레이 역할을 함

- 장점
  1. 인스턴스에서 인터넷에 노출되는 SSH(22) 등의 인바운드 포트를 열 필요가 없음 -> 아웃바운드만 사용
  2. 인스턴스의 SSH key pair의 생성·배포·관리 불필요 -> aws IAM 인증정보로 접속
  3. 인스턴스 내 사용자 계정과 비밀번호 관리 불필요 -> ssm-user 라고 하는 SSM 전용 계정 사용
  4. 세션 중 모든 명령과 그 출력을 로깅 가능 -> 민감정보는 제외 가능
