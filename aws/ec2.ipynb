# EC2 서비스

- EC2(Elastic Compute Cloud) 서비스

## 인스턴스 가동

- EC2 > Instances > Launch an instance

- 가동시 설정 사항
  1. Name and Tags
  2. Application and OS Images (Amazon Machine Image)
  3. Instance type
  4. Key pair (login)
  5. Network settings
  6. Configure storage
  7. Advanced details

## 인스턴스 유형

- 명명규칙: 메인 카테고리 + 세대 + 옵션 카테고리
  - 예시: t4g = t(T 패밀리) + 4(4세대) + g(ARM64 기반)

- 메인 카테고리
  - T 패밀리: 저가형, 순간확장(burstable)
  - M 패밀리: 일반목적
  - C 패밀리: 고성능 CPU
  - R 패밀리: 대용량 메모리 
  - X 패밀리: 초대용량 메모리 
  - D 패밀리: 대용량 HDD 저장장치
  - I 패밀리: 대용량 SDD 저장장치
  - P, G, CG 패밀리: GPU 장착
  - F 패밀리: FPGA 장착

- 옵션 카테고리
  - g: ARM64 기반
  - d: 로컬 NVMe SSD
  - n: 네트워크/EBS 최적화

- vCPU 및 메모리 크기
  - nano
  - micro
  - small
  - large
  - x2large
