# AWS 시스템 매니저

- 인스턴스 내부에 미리 시스템 에이전트를 설차하여 가동중이고
- 해당 에이전트를 사용하여 접속/패치관리 등 다양한 작업 수행
- 제공하는 도구
  - 노드 도구
    - 세션 관리자
    - 상태 관리자
    - 원격 명령 실행
    - 패치 관리자
  - 변경관리 도구
    - 도큐먼트
    - 변경 관리자


## 세션 관리자

- ssh 연결을 대체하는 터미널 연결 도구
  - ssh 등을 사용하여 외부에서 인스턴스로 직접 인바운드 접속을 하는 것이 아니라
  - 인스턴스 내에 SSM 에이전트가 설치되어 SSM 서비스로 아웃바운드 접속하고
  - 사용자는 SSM 서비스로 접속하여 SSM 서비스가 중간에서 릴레이 역할을 함
- 장점
  - 인스턴스에서 인터넷에 노출되는 SSH(22) 등의 인바운드 포트를 열 필요가 없음 -> 아웃바운드만 사용
  - 인스턴스의 SSH key pair의 생성·배포·관리 불필요 -> aws IAM 인증정보로 접속
  - 인스턴스 내 사용자 계정과 비밀번호 관리 불필요 -> ssm-user 라고 하는 SSM 전용 계정 사용
  - 세션 중 모든 명령과 그 출력을 로깅 가능 -> 민감정보는 제외 가능
