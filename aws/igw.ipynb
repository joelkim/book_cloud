# 인터넷 게이트웨이와 NAT

- VPC내의 VM 인스턴스가 외부 인터넷 망과 연결하려면 다음 두 가지 방법을 사용가능
  - 인터넷 게이트웨이(IGW: Internet Gateway) 단독 사용
  - 인터넷 게이트웨이 와 NAT(Network Address Translation) 서버 동시 사용

## 인터넷 게이트웨이

- 인터넷 게이트웨이는 VPC 내부가 아니라 AWS가 독자적으로 운영하는 가상 리소스

### 아이피가 없음

- 인터넷 게이트웨이는 자신만의 사설 아이피와 퍼블릭 아이피를 가지지 않음
- 사설 아이피가 없으므로 인스턴스가 가진 OS 레벨 라우팅 테이블로는 게이트웨이로 라우팅 불가
- 아이피가 아닌 리소스 아이디로 라우팅하는 서브넷 레벨 라우팅 테이블로만 가능
- 따라서 서브넷 단위로 인터넷 게이트웨이와 연결됨

### 연결 방식

- 인터넷 게이트웨이는 사설 아이피 뿐 아니라 퍼블릭 아이피도 없음
- 따라서 외부 인터넷에서 아마존 VPC내의 VM 인스턴스로 연결할 때 인터넷 게이트웨이의 아이피로 우선 요청하고 내부로 라우팅되는 방식이 아님
- 연결하고자 하는 하나의 VM 인스턴스가 하나의 퍼블릭 아이피를 할당받고
- 인터넷 게이트웨이는 실제로는 연결된 서브넷내의 모든 퍼블릭 아이피를 가지게 됨
- 외부 인터넷에서 해당 퍼블릭 아이피로 요청이 들어오면 인터넷 게이트웨이가 받아서 1대1 NAT를 통해 내부의 타겟 인스턴스로 라우팅 
- 따라서 인터넷 게이트웨이와 연결된 서브넷 내부에 있더라도 퍼블릭 아이피가 없는 인스턴스는 인터넷 연결 불가

## NAT 서버

- NAT 서버는 VPC내에서 아이피를 차지하는 내부 리소스
